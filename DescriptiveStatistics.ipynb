```python
def convertDate(date):
    #convert only date
    return datetime.datetime.strptime(date,"%d%b%Y")

def convert(date):
    #convert date and time
    return datetime.datetime.strptime(date,"%d%b%Y%H:%M:%S")

def convertTime(date):
    #convert only time
    return datetime.datetime.strptime(date,"%H:%M:%S")

def getMarketBasketDct(d,f):
    #Return nested dct {household:{date:[(item, time) for item in that market basket]}}
    dct = defaultdict(lambda : defaultdict(list))
    with open(d+f) as f:
        header = next(f).split("|")
        hh_id,item = header.index('hh_id'),header.index('item_id'),
        date,time = header.index('trans_date'),header.index('trans_time')
        
        for line in f:
            line = line.split("|")
            dct[line[hh_id]][line[date]]+=[(line[item],line[time])] 
    return dct

marketBaskets = getMarketBasketDct(d,f1)
```

**Get descriptive statistics about baskets**

```python
def averageBasketSize(dct):
    #Returns number of baskets, number of items, items/baskets, and 
    #lists of items/household and unique items/household
    baskets,items,item_counts,unique_items = 0,0,[],[]
    for household in dct.keys():
        basketItems,basket_count = [], len(dct[household].keys())
        for basket in dct[household].keys():
            basketItems += dct[household][basket]
        baskets += basket_count
        items += len(basketItems)
        item_counts += [len(basketItems)/basket_count]
        unique_items += [len(list(set(basketItems)))/basket_count]
    return baskets,items,items/baskets,item_counts,unique_items

baskets,items,avgItemsPerBasket,item_counts,unique_items = averageBasketSize(marketBaskets)
```

**Time between market baskets**

```python
def timeBtwnPurchases(dct):
    #For each household, get the time between market baskets
    timesBetween = []
    for household in dct.keys():
        times = []
        for basket in dct[household].keys():
            times += [convert(basket + dct[household][basket][0][1])]
        times.sort()
        i=0
        while i < (len(times)-1):
            timesBetween.append(times[i+1]-times[i])
            i+=1
    return timesBetween
#Get descriptive statistics - average basket size, average num items/basket, average time/purchases
```

```python
#Frequencies of items
#Frequencies of items by cluster
def load_product_def(pname):
    dct = defaultdict()
    with open(pname) as f:
        next(f)
        for line in f:
            line = line.split("|")
            dct[line[0]] = [".".join(line[1].split(" "))]
    return dct

def load_product_def_consolidated(pname):
    #load conversions for product table
    #Altered to group subcategories
    dct = defaultdict()
    with open(pname) as f:
        next(f)
        for line in f:
            line = line.split("|")
            if line[5] == "CONVENIENCE":#Added to consolidate
                dct[line[0]] = ["CONVENIENCE"]
            elif re.match(r"""[0-9\-]+""",line[3]):
                dct[line[0]] = ["CARDS.RENTALS"]
            else:
                dct[line[0]] = [".".join(line[1].split(" "))]
    return dct

def load_product_def_truncated(pname):
    #load conversions for product table
    #Altered to group subcategories
    dct = defaultdict()
    with open(pname) as f:
        next(f)
        for line in f:
            line = line.split("|")
            if line[5] == "CONVENIENCE":#Added to consolidate
                pass
                #dct[line[0]] = ["CONVENIENCE"]
            elif re.match(r"""[0-9\-]+""",line[3]):
                pass
                #dct[line[0]] = ["CARDS.RENTALS"]
            else:
                dct[line[0]] = [".".join(line[1].split(" "))]
    return dct

pnames = load_product_def_truncated(pdef)
```

```python
def countItems(marketBaskets,pnames):
    c = Counter()
    keys = set(pnames.keys())
    for household in marketBaskets.keys():
        for basket in marketBaskets[household].keys():
            for item in marketBaskets[household][basket]:
                if item[0] in keys:
                    c[pnames[item[0]][0]] +=1
    return nltk.FreqDist(c)

allItems = countItems(marketBaskets,pnames)
```
