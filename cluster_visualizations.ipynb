```python
def matchLDA(fname,numtopics,pnames):
    #"LDA_50_topics_100_chunks_50_passes_SIMPLIFIED.txt"
    #Finds all items in category that match the tile cluster
    lda = gensim.models.LdaModel.load(fname)
    cleanedTopics = [getSentences(topic,False) for topic in lda.show_topics(numtopics,num_words=60)]
    return [item for item in matchTopics(cleanedTopics,pnames)]

def writeAllLDAToFile(files,categories):
    dirname = ldaModelsFolder
    for f in files:
        match_n = matchLDA(dirname+f,getFileInfo(f),categories)
        match_n.sort(reverse=True)
        pickle.dump(match_n,open(dirname+f[:-10]+"_saved_Topics.txt",'w'))
    return

 

writeAllLDAToFile(files[1:],tileCategories)
```

```python
def getCount(dirname,fname): 
    #Returns {'Tile':counter(),'Grout',counter()} etc
    tileCategories = loadCategories()
    tileItems = set(tileCategories.keys())
    counts = defaultdict(Counter)
    numtopics = getFileInfo(fname)
    lda = gensim.models.LdaModel.load(dirname+fname)
    cleanedTopics = [getSentences(topic) for topic in lda.show_topics(numtopics,num_words=60)]
    for sentence in cleanedTopics:
        for word in sentence:
            if word in tileItems:
                counts[tileCategories[word]][word] +=1
    return counts
```
