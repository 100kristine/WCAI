**Helpers**

```python
def openProductsFile(name="Product.txt"):
    #Open products file
    return [line.split("|") for line in open(name).read().split("\n")]

def getLinesByTopic(lines,topic="FLOOR AND WALL TILE",field=5):
    #Get all lines 
    #TILE SET MATERIALS/TOOLS
    return [line for line in lines if len(line)>field and line[field]==topic]

def getLinesByItem(items):
    #Pull out all lines that contain the item names in their description
    topics = getLinesByTopic(lines,topic="TILE SET MATERIALS/TOOLS",field=5)
    topics += getLinesByTopic(lines,topic="FLOOR AND WALL TILE",field=5)
    keep = []
    for line in topics:
        for word in items:
            if word in line[1]:
                keep.append(".".join(line[1].split(" ")))
                break
    return keep
```

For each topic generated by the LDA, see the intersection between that topic and items. 

matchLDA will clean the topics and check them against list of tools provided

```python
def cleanEntry(topic):
    return [i.split('*')[1] for i in topic.split(" + ")]

def cleanTopic(topic):
    #Separate topics into readable lines 
    allItems = []
    for sent in topic:
        allItems += cleanEntry(sent)
    return allItems

def getSentences(topic):
    #Returns list of readable sentences
    return cleanEntry(topic)

def matchTopics(topics,tools,items):
    def getMatch(topic,items):
        #find the number of in the intersection between topic and items
        #Return as tuples (percentage,topic)
        match = set(topic) & set(tools)
        return len(match),match
    def countCategories(topic,items):
        #Count main categories that the items fall into
        count,match = 0,[]
        for item in items:
            for word in topic:
                if item in word:
                    count += 1
                    match += [word]
                    break
        return count,match
    return [(getMatch(topic,items),countCategories(topic,items),topic) for topic in topics]

def matchLDA(fname,numtopics,tools,items):
    #"LDA_50_topics_100_chunks_50_passes_SIMPLIFIED.txt"
    lda = gensim.models.LdaModel.load(fname)
    cleanedTopics = [getSentences(topic) for topic in lda.show_topics(numtopics,num_words=40)]
    return [item for item in matchTopics(cleanedTopics,tools,items)]
```

```python
#matches = matchLDA("LDA_100_topics_2000_chunks_50_passes.txt",100,tools)
#matches.sort(reverse=True)

def saveFile(fname,numtopics,tools,items):
    matches = matchLDA(fname,numtopics,tools,items)
    matches.sort(reverse=True)
    f = open("Toy_Cluster_top_matches"+fname[:-4]+".txt",'w')
    for match in matches[:5]:
        c1 = str(match[0][0])
        match1 = " ".join(list(match[0][1])) if len(match[0][1]) > 0 else " "
        c2 = str(match[1][0])
        match2 = " ".join(list(match[1][1])) if len(match[1][1]) > 0 else " "
        f.writelines("\n".join([c1 + " item matches", match1, c2 + " category matches",match2]))
        f.writelines("\n".join(match[2])+"\n\n")
    f.close()
    return
```
